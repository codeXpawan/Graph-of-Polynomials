```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Taking INPUTS
degree_of_polynomial = int(input('Degree of polynomial (It must be integer) : '))
constant = float(input('Constant (It should be a real number) : '))
```

```python
#Taking coefficient for every polynomials
coefficient_of_polynomials = np.zeros(degree_of_polynomial)
for i in range(degree_of_polynomial):
  coefficient_of_polynomials[i] = float(input('Coefficent of x^{}: '.format(i+1)))
```

```python
#define the value for the max range of x
x = int(input('x (range from -ve to +ve): '))
array = np.arange(-x,x+1)
x = np.tile(array,degree_of_polynomial).reshape((degree_of_polynomial,x*2+1)).transpose()
for i in range(degree_of_polynomial):
  x[:,i] = x[:,i]**(i+1)
# print(x)
```

```python
coefficient_of_polynomials = np.reshape(coefficient_of_polynomials,(degree_of_polynomial,1))
# print(coefficient_of_polynomials)
```

```python
y = np.matmul(x,coefficient_of_polynomials) + constant
# print(y)
```

```python
plt.plot(array,y)
plt.xlabel('x_axis')
plt.ylabel('y_axis')
plt.title('graph of polynomial')
plt.show()
```
